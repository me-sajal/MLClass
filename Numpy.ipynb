# Diffrences between python list and numpy array
* Numpy array are memory efficient than python list
* Nympy array are faster than python list
* Numpy array supports homogenous data type but python supports multiple data types.


```python
np.random.randint(1,10) #generates random integer on range [1, 10] where 10 doesnot exits
```

```python
# If we want generate same random number then np.random.seed(seed_value)
np.random.seed(55)
np.random.randint(1,10)
```

```python
x = np.array([2, 10, 20, 50, 100], dtype = np.int8) #control dtype datatype and memory
x
```

```python
x = np.array([2, 10, 20, 50, 100], dtype = np.int16) # we can control datatype -> memory efficient
x
```

```python
np.random.seed(40)
list1 =[np.random.randint(1,10) for i in range (50000)]
print(list1)
```

```python
%%time
sq = [item **2 for item in list1] #slower than array
```

```python
list1_array = np.array(list1) # faster than compreension
```

```python
%%time
sq = list1_array ** 2
```
